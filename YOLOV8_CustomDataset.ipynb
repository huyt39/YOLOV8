```python
#Create training yaml file:
%cd /content/drive/MyDrive
!rm data/mydataset.yaml #neu co file thi xoa
!echo 'train: /content/drive/MyDrive/data/train' >>data/mydataset.yaml
!echo 'val: /content/drive/MyDrive/data/train' >>data/mydataset.yaml
!echo 'nc: 1' >>data/mydataset.yaml #number of class
!echo "names: ['fire']" >>data/mydataset.yaml
```

```python
#Train with CLI:
!yolo task=detect mode=predict model=yolov8n.pt data=data/mydataset.yaml epochs=50 imgsz=640
```

```python
#Train with Python API:
from ultralytics import YOLO

model = YOLO("yolov8n.pt") #load pretrained va finetune vai layer cuoi. Transfer learning

result = model.train(data = 'data/mydataset.yaml', epochs = 30)
```
